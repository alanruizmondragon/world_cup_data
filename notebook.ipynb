Unificamos los datos y obtenemos los schemas

```python
cps_id = pd.concat([cps1, cps2, cps3, cps4, cps5, cps6, cps7, cps8, cps9, cps10, 
                    cps11, cps12, cps13, cps14, cps15, cps16, cps17, cps18, cps19, cps20, 
                    cps21, cps22, cps23, cps24, cps25, cps26, cps27, cps28, cps29, cps30, 
                    cps31, cps32], ignore_index=True)
```

Manipulamos la tabla con nuevos nombres, así como el tipo de las variables llave futuras

```python

cps_id.rename(columns = {'D_mnpio':'municipio', 
                         "d_codigo":"cp", 
                         "d_estado":"estado", 
                         "c_estado":"clave_edo", 
                         "c_mnpio":"clave_mun"}, inplace = True)
```

```python
cps_id.info()
```

```python
cps_id["clave_mun"] = cps_id["clave_mun"].astype(str)
cps_id["clave_edo"] = cps_id["clave_edo"].astype(str)
cps_id["clave_mun"] = cps_id["clave_mun"].str.zfill(3)
```

```python
cps_id.head()
```

Construimos el indenficiador del municipio y eliminamos duplicados

```python
cps_id["municipio_id"] = cps_id["clave_edo"] + cps_id["clave_mun"]

#Convertimos en entero al identificador

cps_id["municipio_id"] = cps_id["municipio_id"].astype(int)

#Mantenemos los campos necesarios
cps_id = cps_id[["municipio_id", "municipio", "cp", "estado"]]

#Eliminamos duplicados para la agrupación
cps_id = cps_id.drop_duplicates()
```

```python
#hacemos left join para quedarnos con los cp

atmcp = atm.merge(cps_id, on = "municipio_id", how ="left")
```
